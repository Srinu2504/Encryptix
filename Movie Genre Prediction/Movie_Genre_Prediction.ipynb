```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score

from google.colab import drive


drive.mount('/content/drive')

# Loading dataset
train_data_path = '/content/drive/My Drive/Movie_Genre/train_data.txt'
test_data_path = '/content/drive/My Drive/Movie_Genre/test_data.txt'
test_data_solution_path = '/content/drive/My Drive/Movie_Genre/test_data_solution.txt'

# Data Preprocessing
def load_train_data(file_path):
    data = pd.read_csv(file_path, delimiter=' ::: ', engine='python', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
    return data

def load_test_data(file_path):
    data = pd.read_csv(file_path, delimiter=' ::: ', engine='python', names=['ID', 'TITLE', 'DESCRIPTION'])
    return data

def load_test_solution(file_path):
    data = pd.read_csv(file_path, delimiter=' ::: ', engine='python', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
    return data

train_df = load_train_data(train_data_path)
test_df = load_test_data(test_data_path)
test_solution_df = load_test_solution(test_data_solution_path)

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s]', '', text)
    return text.strip()

train_df['DESCRIPTION'] = train_df['DESCRIPTION'].apply(preprocess_text)
test_df['DESCRIPTION'] = test_df['DESCRIPTION'].apply(preprocess_text)

# Graphs
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='GENRE', order=train_df['GENRE'].value_counts().index)
plt.title('Genre Distribution in Training Data')
plt.xticks(rotation=45)
plt.show()

# TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)

X_train_tfidf = tfidf.fit_transform(train_df['DESCRIPTION'])
X_test_tfidf = tfidf.transform(test_df['DESCRIPTION'])

# Models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Naive Bayes': MultinomialNB(),
    'Support Vector Machine': SVC()
}

# Train and evaluate each model
for model_name, model in models.items():
    print(f"\nTraining {model_name}...")
    model.fit(X_train_tfidf, train_df['GENRE'])
    y_test_pred = model.predict(X_test_tfidf)

    test_df['PREDICTED_GENRE'] = y_test_pred
    merged_df = test_df.merge(test_solution_df[['ID', 'GENRE']], on='ID', how='left', suffixes=('', '_TRUE'))

    accuracy = accuracy_score(merged_df['GENRE'], merged_df['PREDICTED_GENRE'])
    report = classification_report(merged_df['GENRE'], merged_df['PREDICTED_GENRE'], output_dict=True)
    conf_matrix = confusion_matrix(merged_df['GENRE'], merged_df['PREDICTED_GENRE'])
    precision = precision_score(merged_df['GENRE'], merged_df['PREDICTED_GENRE'], average='weighted')

    print(f"Test Accuracy ({model_name}): {accuracy}")
    print(f"Test Classification Report ({model_name}):\n{classification_report(merged_df['GENRE'], merged_df['PREDICTED_GENRE'])}")
    print(f"Confusion Matrix ({model_name}):\n{conf_matrix}")
    print(f"Precision Score ({model_name}): {precision}")


    print(merged_df[['ID', 'TITLE', 'PREDICTED_GENRE', 'GENRE']])

    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=merged_df['GENRE'].unique(), yticklabels=merged_df['GENRE'].unique())
    plt.title(f'Confusion Matrix ({model_name})')
    plt.xlabel('Predicted Genre')
    plt.ylabel('True Genre')
    plt.show()

    # Plot precision, recall, f1-score
    metrics_df = pd.DataFrame(report).transpose()
    metrics_df = metrics_df.drop(['accuracy', 'macro avg', 'weighted avg'])

    metrics_df[['precision', 'recall', 'f1-score']].plot(kind='bar', figsize=(10, 6))
    plt.title(f'Precision, Recall, F1-Score ({model_name})')
    plt.xlabel('Genre')
    plt.ylabel('Score')
    plt.xticks(rotation=45)
    plt.show()
```
